```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(n):
    epsilon_values = []
    for i in range(n):
        e = np.random.randn()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100)

plt.plot(data, 'b-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def generate_data(n, generator_type):
    epsilon_values = []
    i = 0
    while i < n:
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        epsilon_values.append(e)
        i += 1
    return epsilon_values

data = generate_data(100, 'U')
plt.plot(data, 'b-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(n, generator_type):
    epsilon_values = [generator_type() for i in range(n)]
    return epsilon_values

data = generate_data(100, np.random.uniform)

plt.plot(data, 'b-')
plt.show()
```

```python
# Exercise 1
import numpy as np

def factorial(n):
    product = 1
    for i in range(1,n+1): #when using range, make sure to start at 1, and end at n+1
        product = product*i
    return product

print(factorial(4))
a = np.arange(1,4) #same thing as above
print(a)
vecfactorial = np.vectorize(factorial)
print(vecfactorial(a))

for j in range(1,5):
    print(j)
    print(factorial(j))
```

```python
# Exercise 2
from numpy.random import uniform

def binomial_rv(n, p):
    drawcount = 0
    for i in range(n):
        U = uniform()
        if U < p:
            drawcount +=1
    return drawcount

binomial_rv(10, 0.5)
```

```python
# Exercise 3
# Estimate area using A/r^2
import numpy as np

n = 1000000

count = 0
for i in range(n):
    k, l = np.random.uniform(), np.random.uniform()
    a = np.sqrt((k - 0.5)**2 + (l - 0.5)**2)
    if a < 0.5:
        count +=1
        
area_guess = count / n

print(area_guess * 4)  # dividing by radius**2
```
